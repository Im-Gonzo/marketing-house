```python
import pandas as pd
```

```python
df = pd.read_csv("../src/data/customer/raw/sales_dataset.csv")
```

```python
df.dtypes
```

```python
df.head()
```

```python
rfm_df = df[
    [
        "Customer ID",
        "Order Date",
        "Sales",
    ]
].copy()
```

```python
rfm_df.rename(
    columns={
        "Customer ID": "CustomerID",
        "Order Date": "InvoiceDate",
        "Sales": "Monetary",
    },
    inplace=True,
)
```

```python
rfm_df["CustomerID"] = rfm_df["CustomerID"].astype(str)
rfm_df["InvoiceDate"] = pd.to_datetime(rfm_df["InvoiceDate"], format="%d/%m/%Y")
```

```python
rfm_df.to_csv("../src/data/customer/preprocessed/test.csv", index=False)
```
